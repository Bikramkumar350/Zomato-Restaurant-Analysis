```python
merged = pd.merge(df_data, df_country, on='Country', how='left')
print("after merge sample:\n", merged.head())
```

```python
cols_to_drop = [c for c in ['Unnamed: 0', 'Restaurant ID','Locality Verbose','Country-Code','Longitude','Latitude'] if c in merged.columns]
if cols_to_drop:
    merged.drop(cols_to_drop, axis=1, inplace=True)
print("columns after drop:\n", merged.columns.tolist())
```

```python
print("nulls before dropping:\n", merged.isnull().sum())
```

```python
merged = merged.dropna()
print("nulls after dropping:\n", merged.isnull().sum())
```

```python
print(merged.Country.value_counts())
```

```python
merged.to_csv(r"cleaned_zomato.csv", index=False)
print("cleaned data written to cleaned_zomato.csv")
```
